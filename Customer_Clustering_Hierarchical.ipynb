```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder
lr=LabelEncoder()
d['Annual_Income_(k$)']=lr.fit_transform(d['Annual_Income_(k$)'])
```

```python
x=d['Annual_Income_(k$)']
y=d['Spending_Score']
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(10,7))
dendrogram=sch.dendrogram(sch.linkage(x.values.reshape(-1, 1),method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
ag=AgglomerativeClustering(n_clusters=5,linkage='ward')
y_pred=ag.fit_predict(x.values.reshape(-1,1))
y_pred
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x,y,c=y_pred)
```
